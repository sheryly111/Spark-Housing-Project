This notebook merges property assessment databases from the years 2018 - 2024 and cleans it up for future use.

```python
df = pd.read_csv("raw_property_assessment.csv")
df.head()
```

```python
cols_keep = ["PID", "CM_ID", "GIS_ID", "ST_NUM", "ST_NAME", "UNIT_NUM", "CITY", "ZIP_CODE", "OWNER", "INT_COND", "EXT_COND", "OVERALL_COND", "BDRM_COND", "HEAT_TYPE", "AC_TYPE", "BED_RMS"]

df_new = df[cols_keep] 
```

```python
df_new.head()
```

```python
add_cols = ["ST_NUM", "ST_NAME", "CITY", "ZIP_CODE"]
df_new[add_cols] = df_new[add_cols].fillna("").astype(str)
df_new["full_address"] = df_new["ST_NUM"].str.strip() + " " + df_new["ST_NAME"].str.strip() + " " + df_new["CITY"].str.strip() + " " + df_new["ZIP_CODE"].str.strip()
df_new["full_address"] = df_new["full_address"].str.replace(r"\s+", " ", regex=True).str.strip()

df_cleaned = df_new.drop(add_cols, axis=1)
```

```python
df_cleaned.head()

df_cleaned.to_csv("cleaned_property_assessment.csv", index=False)
```
